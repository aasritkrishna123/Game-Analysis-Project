# Columns in Dataset (Video Games Data)

1. S.No: A Serial Number or unique index for the game entry

2. Title: The name of the video game.

3. Release Date: The date the game was officially released.

4. Team: The developers and/or publishers responsible for the game (typically a list).

5. Rating: The average user or critic rating for the game (a numerical value).

6. Times Listed: The total count of times the game has been included on various "best of" lists

7. Number of Reviews: The total count of reviews the game has received

8. Genres: The categories or types of the game (e.g., 'Adventure', 'RPG', 'Indie')

9. Summary: A brief descriptive text summarizing the game's plot, world, and mechanics.

10. Reviews: A list of individual text reviews for the game.

11. Plays: The total count of users who have marked the game as played/completed.

12. Playing: The total count of users who currently have the game marked as being played.

13. Backlogs: The total count of users who own the game but have it on their "backlog" (yet to play or finish).

14. Wishlist: The total count of users who have added the game to their wishlist.

# Columns in Dataset (vgsales Data)

1. Rank: The ranking of the game based on its total global sales.

2. Name: The name of the video game.

3. Platform: The specific gaming console or system the game was released on (e.g., 'Wii', 'PS4', 'PC').

4. Year: The year the game was released.

5. Genre: The category or type of game (e.g., 'Sports', 'Role-Playing', 'Platform').

6. Publisher: The company responsible for publishing the game.

7. NA_Sales: The sales volume of the game in North America (in millions of units).

8. EU_Sales: The sales volume of the game in Europe (in millions of units).

9. JP_Sales: The sales volume of the game in Japan (in millions of units).

10. Other_Sales: The sales volume of the game in all other regions not covered above (in millions of units).

11. Global_Sales: The total worldwide sales of the game (the sum of NA_Sales, EU_Sales, JP_Sales, and Other_Sales), in millions of units.

# Questions of 'Games' Data

```python
# What are the top-rated games by user reviews?

# Sort games by rating in descending order
top_rated_games = games.sort_values(by='Rating', ascending=False)

# Display the top 10 games by rating
display(top_rated_games[['Title', 'Rating', 'Number of Reviews']].head(10))
```

```python
# Which developers (Teams) have the highest average ratings?

# Calculate the average rating for each team
average_rating_by_team = games.groupby('Team')['Rating'].mean().reset_index()

# Sort teams by average rating in descending order
top_teams_by_rating = average_rating_by_team.sort_values(by='Rating', ascending=False)

# Display the top 10 teams by average rating
display(top_teams_by_rating.head(10))
```

```python
# What are the most common genres in the dataset?

from collections import Counter

# Split the genres string in each row and create a list of all genres
all_genres = [genre for genres_list in games['Genres'].str.split(',') for genre in genres_list if genre]

# Count the occurrences of each genre
genre_counts = Counter(all_genres)

# Get the most common genres
most_common_genres = genre_counts.most_common(10)

# Display the most common genres
print("Most common genres:")
for genre, count in most_common_genres:
    print(f"{genre}: {count}")
```

```python
# Which games have the highest backlog compared to wishlist?

# Convert 'Backlogs' and 'Wishlist' columns to numeric, coercing errors
games['Backlogs'] = pd.to_numeric(games['Backlogs'].str.replace('K', 'e3'), errors='coerce')
games['Wishlist'] = pd.to_numeric(games['Wishlist'].str.replace('K', 'e3'), errors='coerce')

# Calculate the difference between Backlogs and Wishlist
games['Backlog_vs_Wishlist'] = games['Backlogs'] - games['Wishlist']

# Sort games by the difference in descending order
highest_backlog_games = games.sort_values(by='Backlog_vs_Wishlist', ascending=False)

# Display the top 10 games with the highest backlog compared to wishlist
display(highest_backlog_games[['Title', 'Backlogs', 'Wishlist', 'Backlog_vs_Wishlist']].head(10))
```

```python
# What is the game release trend across years?

# Extract the year from the 'Release Date' column
games['Release_Year'] = pd.to_datetime(games['Release Date'], errors='coerce').dt.year

# Count the number of games released each year
release_counts_by_year = games['Release_Year'].value_counts().sort_index()

# Display the release counts by year
display(release_counts_by_year)
```

```python
# What is the distribution of user ratings?

import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram of the 'Rating' column
plt.figure(figsize=(10, 6))
sns.histplot(games['Rating'].dropna(), bins=20, kde=True)
plt.title('Distribution of User Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Sort games by wishlist in descending order
top_wishlisted_games = games.sort_values(by='Wishlist', ascending=False)

# Display the top 10 most wishlisted games
display(top_wishlisted_games[['Title', 'Wishlist']].head(10))
```

```python
# Whats the average number of plays per genre?

# Convert 'Plays' column to numeric, coercing errors
games['Plays'] = pd.to_numeric(games['Plays'].astype(str).str.replace('K', 'e3'), errors='coerce')

# Split genres and explode the dataframe
games_exploded = games.assign(Genres=games['Genres'].str.split(',')).explode('Genres')

# Calculate the average plays per genre
average_plays_by_genre = games_exploded.groupby('Genres')['Plays'].mean().reset_index()

# Sort by average plays in descending order and display
display(average_plays_by_genre.sort_values(by='Plays', ascending=False))
```
